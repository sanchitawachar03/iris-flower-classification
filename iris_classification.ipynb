```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_excel("iris.train.xlsx")
```

```python
data.head()
```

```python
x = data.iloc[:,:-1]
```

```python
y = data.iloc[:,-1]
```

```python
y
```

```python
model = DecisionTreeClassifier()
```

```python
model.fit(x.values,y.values)
```

```python
predictions = model.predict([[5.7,3.8,1.7,0.3]])
```

```python
predictions 
```

```python
data_test = pd.read_excel("iris.test.xlsx")
```

```python
x_test = data_test.iloc[:,:-1]
```

```python
predictions_all = model.predict(x_test.values)
```
